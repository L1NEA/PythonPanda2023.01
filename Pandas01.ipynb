```python
# Importando Biblioteca Pandas no Python (não é uma biblioteca nativa do Python; o Colab ja tem ela instalada por padrão, mas em um IDE Local precisa instalar a biblioteca)

import pandas as pd

combustiveis_df = pd.read_excel("ca-2021-02.xlsx")
```

```python
# Comand Dataframe.shape

print (combustiveis_df.shape)
# Saída = número de linhas, número de colunas

print (combustiveis_df.shape[0])
# Saída = só o número de linhas

print (combustiveis_df.shape[1])
# Saída = só o número de colunas
```

```python
# Mostra as colunas (nomes) e que tipo de dados cada uma tem

print (combustiveis_df.info())
```

```python
# Comando Dataframe.describe()

display (combustiveis_df.describe())

# Saída = estatisticas mais básicas do Dataframe (somente informações numéricas) (Contagem de linhas, média, desvio padrão, "parcelas" e máximo)
```

```python
# Filtrar apenas por uma coluna

display(combustiveis_df['Revenda'])
```

```python
# Criar um novo Dataframe apenas com as colunas que eu quero de outro Dataframe

ca_df = combustiveis_df[['Revenda','Municipio','Produto','Valor de Venda']]
display (ca_df)
```

```python
# Exibe a 4ª linha

display (ca_df.loc[3])
```

```python
# Exibir da 10ª linha até a 20ª linha

display(ca_df.loc[9:19])
```

```python
# Exibir da 2ª linha até a 10ª pulando de 2 em 2

display(ca_df.loc[2:10:2])
```

```python
# Criar um dataframe gas_df contendo apenas as 4 colunas (Revenda, Municipio, Produto, Valor de Venda) somente com combustivel sendo GASOLINA e exibir na tela

gas_df = ca_df.loc[ca_df['Produto'] == 'GASOLINA']
display (gas_df)
```

```python
# Mostrar o valor da gasolina mais cara

display (gas_df['Valor de Venda'].max())
```

```python
# Mostrar posto e municipio da gasolina mais cara

display(gas_df[['Revenda', 'Municipio', 'Valor de Venda']].max())
```

```python
# Dataframe.loc[] com multiplas condições para filtragem

# Quais são os preços dos postos que vendem ETANOL na cidade de SANTO ANDRE ordenado do menor -> maior

etanol_santoandre_df = ca_df.loc[(ca_df ['Produto'] == 'ETANOL') & (ca_df ['Municipio'] == 'SANTO ANDRE')]
display (etanol_santoandre_df.sort_values(by='Valor de Venda'))
```

```python
# Qual é a média de preço dos combustíveis GASOLINA e GASOLINA ADITIVADA do bairro VILA FORMOSA em SÃO PAULO
display (combustiveis_df.loc[(combustiveis_df['Bairro'] == 'VILA FORMOSA') &
                             (combustiveis_df ['Municipio'] == 'SAO PAULO') &
                             ((combustiveis_df ['Produto'] == 'GASOLINA') | (combustiveis_df ['Produto'] == 'GASOLINA ADITIVADA')),
                             ['Valor de Venda']].mean())
```

```python
# Como mostrar média de valor POR COMBUSTÍVEL

media_por_combustivel_df = ca_df[['Produto','Valor de Venda']].groupby(by= 'Produto').mean().round(2)
display(media_por_combustivel_df)
```

```python
# Adicionar uma coluna de valor booleano no combustiveis_df chamada "Ativo" que, por padrão, vai ser True para TODAS as linhas

combustiveis_df ['Ativo'] = True
print (combustiveis_df.info())
display (combustiveis_df.head())
```

```python
# Criar uma coluna nova chamada "Obs" que preencha automaticamente "MELHOR CIDADE" quando o municipio for SAO PAULO

combustiveis_df['Obs'] = ["Melhor Cidade" if municipio == 'SAO PAULO' else '' for municipio in combustiveis_df['Municipio']]
print (combustiveis_df.info())
display (combustiveis_df)
```

```python
# Exportar para excel o dataframecom etanol em Santo Andre
etanol_santoandre_df.to_excel('etanol_santoandre.xlsx', sheet_name = 'Etanol em Santo Andre')
```
