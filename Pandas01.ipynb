```python
# Como mostrar média de valor POR COMBUSTÍVEL

media_por_combustivel_df = ca_df[['Produto','Valor de Venda']].groupby(by= 'Produto').mean().round(2)
display(media_por_combustivel_df)
```

# **Calcular o número de postos de gasolina por habitantes temos na amostragem de combustiveis nov/2021**

```python
# Renomear colunas

num_habitantes_df = pd.read_csv("ibge_num_habitantes_estimado.csv", sep=";")
num_habitantes_df.rename (columns={"Estado" : "Estado - Sigla"}, inplace=True)
display (num_habitantes_df)
```

```python
# Fazer um MERGE de dois dataframes em outro dataframe

merge_df = combustiveis_df.merge (num_habitantes_df, how="inner", on=['Municipio', 'Estado - Sigla'])
display (merge_df)
print(merge_df.info())
```

```python
#Destruir coluna completamente vazia (todas as linhas são nulas)

merge_df.dropna(axis='columns', inplace=True)
print(merge_df.info())
```

```python
#Destruir colunas que estiverem completamente vazias (todas as linhas são nulas)

merge_df.dropna(axis='columns', inplace=True)
print(merge_df.info())
```

```python
# Excluir colunas selecionadas

colunas=['Regiao - Sigla', 'Nome da Rua', 'Numero Rua', 
         'Bairro', 'Cep', 'Produto', 'Data da Coleta', 'Valor de Venda',
         'Unidade de Medida', 'Bandeira', 'Ativo', 'Status - Valor de Venda', 'Obs']

merge_df.drop(labels=colunas, axis=1, inplace=True)
print(merge_df.info())
display (merge_df)
```

```python
# Destruir linhas duplicadas

merge_df.drop_duplicates (inplace = True)
display (merge_df.head(100))
```

```python
# Agrupar e contar quantos postos tem na cidade

postos_por_municipio_df = merge_df.groupby(by=['Estado - Sigla', 'Municipio', 'NumHabitantes2021']).count() #agrupa as informações iguais nessas colunas e conta quantos itens tem
postos_por_municipio_df.drop('CNPJ da Revenda', axis=1, inplace=True) #descartou uma coluna
postos_por_municipio_df.reset_index(inplace=True)
postos_por_municipio_df.rename(columns={"Revenda": "Número de Postos"}, inplace=True) #renomeou uma coluna para trazer a informação desejada
display(postos_por_municipio_df.info())
display(postos_por_municipio_df)
```

```python
# FINAL - **Calcular o número de postos de gasolina por habitantes temos na amostragem de combustiveis nov/2021**

postos_por_municipio_df ['Postos por Habitantes'] = postos_por_municipio_df['Número de Postos'] / postos_por_municipio_df['NumHabitantes2021']
display (postos_por_municipio_df)
```
