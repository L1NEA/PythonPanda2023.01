# **ANÁLISE**

```python
import pandas as pd
```

```python
#Importar o dataframe

medicamentos_df = pd.read_excel("TA_PRECO_MEDICAMENTO.xlsx")
```

```python
display(medicamentos_df)
```

```python
print (medicamentos_df.info())
```

```python
# Filtrar as colunas necessárias para análise

med_df = medicamentos_df [['SUBSTÂNCIA', 'LABORATÓRIO', 'PRODUTO', 'APRESENTAÇÃO', 'CLASSE TERAPÊUTICA',
                           'TIPO DE PRODUTO (STATUS DO PRODUTO)', 'REGIME DE PREÇO', 'PF Sem Impostos', 'RESTRIÇÃO HOSPITALAR', 'TARJA']]
display(med_df)
```

```python
# Criar um dataframe que exibe somente medicamentos Tarja Preta

tarja_preta_df = med_df.loc[(med_df['TARJA'] == 'Tarja Preta') | (med_df['TARJA'] == 'Tarja Preta (**)')]
display (tarja_preta_df)
```

```python
# Mostrar o medicamento em tarja preta mais caro e suas informações

display(tarja_preta_df[['LABORATÓRIO', 'PRODUTO', 'CLASSE TERAPÊUTICA', 'RESTRIÇÃO HOSPITALAR', 'PF Sem Impostos']].max())
```

```python
# Agrupar e contar quais CNPJs vendem medicamentos de cada tipo de tarja

med_por_tarja_df = med_df[['LABORATÓRIO', 'TARJA']].groupby(by=['LABORATÓRIO', 'TARJA']).count()
display(med_por_tarja_df)
```

```python
# Extrair planilha de quais CNPJs vendem medicamentos de cada tipo de tarja

med_por_tarja_df.to_excel('med_por_tarja.xlsx', sheet_name = 'Medicamentos por tarja - Locais de venda')
```

# **GRÁFICO**

```python
import matplotlib.pyplot as plt
```

```python
# Quantidade de medicamentos por tarja

remedios_restrição_df = med_df[['TARJA', 'RESTRIÇÃO HOSPITALAR']].groupby(by=['TARJA']).count()
remedios_restrição_df.rename(columns={"RESTRIÇÃO HOSPITALAR": "Quant. de Medicamentos"}, inplace=True)
display(remedios_restrição_df)
```

```python
import seaborn as sns

# Definir a áres do gráfico
plt.figure(figsize=(25,15))

# Configurar o gráfico

remedios_restrição_df_grafico = remedios_restrição_df.plot(
    kind="barh",
    ylabel="Tarja do Medicamento",
    xlabel="Quantidade de medicamentos da base",
    title="Quant. de medicamentos de acordo com o tipo de tarja",
    color="red",
    alpha=0.7 #opacidade
)

remedios_restrição_df_grafico.set_ylabel("Tarja do Medicamento")
remedios_restrição_df_grafico.set_xlabel("Quantidade de medicamentos da base")

# Grid
plt.grid(axis="x", alpha=0.7) #grade baseada no eixo x

# Remover as bordas do gráfico
sns.despine()

# Exibir o gráfico
plt.show()
```
